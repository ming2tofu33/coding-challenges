# `92. [대표 유형] 우선순위 큐`

```python
from heapq import heappush, heappop

def solution(queries):
    heap = []
    result = []

    for x in queries:
        if x:
            heappush(heap, x)
        else:
            if heap:
                result.append(heappop(heap))
            else:
                result.append(0)
    
    return result
```

```python
solution([4,10,7,0,0,5,0,15,0])
```

```python
solution([5,5,5,0,3,0,0])
```

```python
solution([0,1,0,0])
```
