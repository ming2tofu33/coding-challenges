# `86. [대표 유형] 최소 신장 트리(Minimum Spanning Tree)`

```python
from heapq import heappush, heappop

def solution(n, edges):
    INF = float('inf')
    dist = [INF] * (n + 1)
    adj = [[] for _ in range(n + 1)]

    for x, y, z in edges:
        adj[x].append([z, y])
        adj[y].append([z, x])
    
    heap = [[0, 1]]
    dist[1] = 0
    visited = {1}

    while heap:
        w, node = heappop(heap)
        visited.add(node)

        if dist[node] < w:
            continue
        
        for nxt_dist, nxt in adj[node]:
            if nxt not in visited and dist[nxt] > nxt_dist:
                dist[nxt] = nxt_dist
                heappush(heap, [nxt_dist, nxt])

    return sum(dist[1:])
```

```python
solution(n=6, edges=[[4,3,4],[1,2,3],[4,6,6],[1,4,10],[5,6,4],[3,2,2],[2,4,5],[6,3,1]])
```
