# `25. 모든 약수의 총합 구하기`

```python
def solution(number):
    add = []
    div = []

    for i in range(1, number + 1):
        for j in range(1, i + 1):
            if i % j == 0:
                div.append(j)
        add.append(sum(div))
        div = []

    return sum(add)
```

```python
solution(5)
```

```python
solution(10)
```
