# `82. [대표 유형] 큐`

```python
from collections import deque

def solution(queries):
    answer = []
    queue = deque()

    for q in queries:
        if q.startswith("push"):
            _, num = q.split()
            queue.append(int(num))
        elif q == "pop":
            answer.append(queue.popleft() if queue else 0)
        elif q == "peek" or q == "front":
            answer.append(queue[0] if queue else 0)
        elif q == "size" or q == "len":
            answer.append(len(queue))
        elif q == "empty":
            answer.append(0 if queue else 1)

    return answer
```

```python
solution(["push 4","push 10","push 7","pop","pop","push 5","pop","push 15","pop"])
```

```python
solution(["push 5","push 5","push 5","pop","push 3","pop","pop"])
```

```python
solution(["pop","push 1","pop","pop"])
```
