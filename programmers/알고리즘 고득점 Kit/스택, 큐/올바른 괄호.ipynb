```python
def solution(s):
    stack = 0
    
    for p in s:
        if p == '(':
            stack += 1
        else:
            if stack:
                stack -= 1
            else:
                return False
    
    return False if stack else True
```

```python
solution("()()")
```

```python
solution("(())()")
```

```python
solution(")()(")
```
