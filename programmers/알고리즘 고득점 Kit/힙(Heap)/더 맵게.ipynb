```python
from heapq import heapify, heappush, heappop

def solution(scoville, K): 
    heapify(scoville)
    cnt = 0
    
    while len(scoville) > 1 and scoville[0] < K:
        low1 = heappop(scoville)
        low2 = heappop(scoville)
        heappush(scoville, low1 + low2 * 2)
        cnt += 1
    
    return -1 if len(scoville) == 1 and scoville[0] < K else cnt
```

```python
solution([1, 2, 3, 9, 10, 12], 7)
```
