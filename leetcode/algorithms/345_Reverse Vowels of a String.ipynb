# `345_Reverse Vowels of a String`

```python
class Solution:
    def reverseVowels(self, s: str) -> str:
        l = 0
        r = len(s) - 1

        vowels = "aeiouAEIOU"
        strlist = list(s)

        while l < r:
            if strlist[l] not in vowels:
                l += 1
            if strlist[r] not in vowels:
                r -= 1
            if strlist[l] in vowels and strlist[r] in vowels:
                strlist[l], strlist[r] = strlist[r], strlist[l]
                l += 1
                r -= 1

        return ''.join(strlist)
```

```python
Solution().reverseVowels("IceCreAm")
```
