# `17_Letter Combinations of a Phone Number`

```python
digits = "23"
#  ["ad","ae","af","bd","be","bf","cd","ce","cf"]

from itertools import combinations

alphabets = {'2': ['a', 'b', 'c'],
             '3': ['d', 'e', 'f'],
             '4': ['g', 'h', 'i'],
             '5': ['j', 'k', 'l'],
             '6': ['m', 'n', 'o'],
             '7': ['p', 'q', 'r'],
             '8': ['t', 'u', 'v'],
             '9': ['w', 'x', 'y', 'z'],
             }

letters = []

for digit in digits:
    letters += alphabets[digit]

comb = []
for c in combinations(letters, len(digits)):
    comb.append(''.join(c))

print(comb)
```
