# `344_Reverse String`

```python
class Solution:
    def reverseString(self, s: list[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        s.reverse()
```

```python
Solution().reverseString(s = ["h","e","l","l","o"])
```

```python
Solution().reverseString(["H","a","n","n","a","h"])
```
