# `509_Fibonacci Number`

```python
class Solution:
    def fib(self, n: int) -> int:
        def f(n):
            if n == 0:
                return 0
            if n == 1:
                return 1
            return f(n - 1) + f(n - 2)

        return f(n)
```

```python
class Solution:
    def fib(self, n: int) -> int:
        if n < 2:
            return n
        
        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return b
```
