# `830_Positions of Large Groups`

```python
class Solution:
    def largeGroupPositions(self, s: str) -> list[list[int]]:
        s += ' '
        
        groups = []
        i, j, N = 0, 0, len(s)

        while j < N:
            if s[i] == s[j]:
                j += 1
            else:
                if j - i >= 3:
                    groups.append([i, j - 1])
                i = j

        return groups
```

```python
s = "abcdddeeeeaabbbcd"

s += ' '

groups = []
i, j, N = 0, 0, len(s)

while j < N:
    if s[i] == s[j]:
        j += 1
    else:
        if j - i >= 3:
            groups.append([i, j - 1])
        i = j
        
print(groups)
```
