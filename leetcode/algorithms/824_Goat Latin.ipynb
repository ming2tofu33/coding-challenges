# `824_Goat Latin`

```python
class Solution:
    def toGoatLatin(self, sentence: str) -> str:
        words = sentence.split()
        vowels = set('aeiouAEIOU')

        for idx, word in enumerate(words):
            if word[0] in vowels:
                words[idx] += 'ma'
            else:
                words[idx] = word[1:] + word[0] + 'ma'
            words[idx] += 'a' * (idx + 1)

        return ' '.join(words)
```
