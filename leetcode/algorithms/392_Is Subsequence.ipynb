# `392_Is Subsequence.ipynb`

```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        i = j = 0

        while i < len(s) and j < len(t):
            if s[i] == t[j]:
                i += 1
            j += 1
            if i == len(s):
                return True
            
        return i == len(s)
```

```python
Solution().isSubsequence(s = "abc", t = "ahbgdc")
```

```python
Solution().isSubsequence(s = "axc", t = "ahbgdc")
```
