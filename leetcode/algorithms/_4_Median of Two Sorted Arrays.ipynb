```python
nums1 = [1,3]
nums2 = [2]

nums1 = [1,2]
nums2 = [3,4]

i, j = 0, 0
merged_len = len(nums1) + len(nums2)
mid = merged_len // 2

while i + j < mid:
    if i < len(nums1) and (j >= len(nums2) or nums1[i] < nums2[j]):
        i += 1
    else:
        j += 1    
```
