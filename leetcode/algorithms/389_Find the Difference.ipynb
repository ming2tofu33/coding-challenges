# `389_Find the Difference`

```python
from collections import Counter

class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        return (Counter(t) - Counter(s)).popitem()[0]
```

```python
Solution().findTheDifference("abcd", "abcde")
```

```python
Solution().findTheDifference(s = "", t = "y")
```
