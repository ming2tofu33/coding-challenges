# `744_Find Smallest Letter Greater Than Target`

```python
class Solution:
    def nextGreatestLetter(self, letters: list[str], target: str) -> str:
        s = set(letters)

        for i in range(ord(target) + 1, 123):
            if chr(i) in s:
                return chr(i)

        return letters[0]
```

```python
Solution().nextGreatestLetter(letters = ["c","f","j"], target = "a")
```

```python
Solution().nextGreatestLetter(letters = ["c","f","j"], target = "c")
```

```python
Solution().nextGreatestLetter(letters = ["x","x","y","y"], target = "z")
```
