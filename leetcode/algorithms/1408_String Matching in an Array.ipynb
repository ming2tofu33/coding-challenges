# `1408_String Matching in an Array`

```python
def stringMatching(words: list[str]) -> list[str]:
    def is_sub_of_another(w: str) -> bool:
        return any(w != other and w in other for other in words)

    return list(filter(is_sub_of_another, words))
```

```python
stringMatching(words = ["mass","as","hero","superhero"])
```

```python
stringMatching(words = ["leetcode","et","code"])
```

```python
stringMatching(words = ["blue","green","bu"])
```
