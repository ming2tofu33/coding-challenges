# `17_Letter Combinations of a Phone Number`

```python
class Solution:
    def letterCombinations(self, digits: str) -> list[str]:
        if not digits:
            return []

        mp = {
            "2": "abc", "3": "def", "4": "ghi", "5": "jkl",
            "6": "mno", "7": "pqrs", "8": "tuv", "9": "wxyz"
        }

        res = []
        path = []

        def dfs(i: int) -> None:
            if i == len(digits):
                res.append("".join(path))
                return
            for ch in mp[digits[i]]:
                path.append(ch)
                dfs(i + 1)
                path.pop()

        dfs(0)
        return res
```

```python
Solution().letterCombinations(digits = "23")
```

```python
Solution().letterCombinations(digits = "2")
```
